```python
cols_to_cap = [
    'Salary',
    'Errors',
    'Assists',
    'PutOuts',
    'CWalks',
    'CRBI',
    'CHmRun',
    'HmRun',        
    'Runs' ,        
    'RBI'  ,       
    'Walks'   ,    
    'Years'    ,   
    'CAtBat'  ,
    'CHits',
    
]

for col in cols_to_cap:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    print(f"{col}: lower={lower_limit}, upper={upper_limit}")

    df[col] = df[col].clip(lower_limit, upper_limit)

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.drop(columns=['Salary']),df['Salary'],test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
alphas = 10**np.linspace(10,-2,100)*0.00001
ridge = Ridge()
coeffs = []
```

```python
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(x_train_sc,y_train)
    coeffs.append(ridge.coef_)
```

```python
from sklearn.metrics import mean_squared_error

train_mse = []
test_mse = []

for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(x_train_sc, y_train)
    
    train_mse.append(mean_squared_error(y_train, ridge.predict(x_train_sc)))
    test_mse.append(mean_squared_error(y_test, ridge.predict(x_test_sc)))
```

```python
plt.figure(figsize=(8,5))
plt.plot(alphas, train_mse, label='Train MSE')
plt.plot(alphas, test_mse, label='Test MSE')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.legend()
plt.title('Bias–Variance Tradeoff (Ridge)')
plt.show()
```

```python
best_alpha = alphas[np.argmin(test_mse)]
print("Best alpha:", best_alpha)
```

```python
ridge_final = Ridge(alpha=best_alpha)
ridge_final.fit(x_train_sc, y_train)
```

```python
y_pred=ridge_final.predict(x_test)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
```

```python
print(mean_squared_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
