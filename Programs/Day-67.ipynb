```python
sns.scatterplot(x=df['cgpa'],y=df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=4)
```
